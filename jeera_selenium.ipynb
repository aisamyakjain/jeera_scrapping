```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service

# Create ChromeOptions object
chrome_options = webdriver.ChromeOptions()

# Add any desired options using add_argument method
chrome_options.add_argument("--disable-dev-shm-usage")  

# Use webdriver_manager to automatically download the appropriate ChromeDriver version
driver = webdriver.Chrome(ChromeDriverManager().install(), options=chrome_options)

# Now you can use the 'driver' object with the specified options

# Example: Open a website
url = "https://www.ncdex.com/products/JEERAUNJHA"
driver.get(url)
```

```python
title = driver.title
print("Title:", title)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(driver.page_source, 'lxml')
```

```python
print(soup.title.text.strip())
```

```python
table  = soup.find("div", class_="scrolltable")

```

```python
list_of_rows=[]
```

```python
for row in table.findAll('tr'):
    
	list_of_cells=[]
	for cell in row.findAll('td'):
     
		list_of_cells.append(cell.get_text(strip=True))
		list_of_rows.append(list_of_cells)
```

```python
import csv
```

```python
#writing to csv
outfile=open('C:/Users/Samiyak Jain/Desktop/Zeera.csv','w')
writer=csv.writer(outfile)
writer.writerows(list_of_rows)
outfile.close()
```

```python
#Showing the file content

with open('C:/Users/Samiyak Jain/Desktop/Zeera.csv', 'r') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        print(row)
```
